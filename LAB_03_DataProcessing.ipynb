```python
#Creating dataframe1
df1 = pd.DataFrame({
        'Name':['Jeevan', 'Raavan', 'Geeta', 'Bheem'],
        'Age':[25, 24, 52, 40],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']})
df1
```

```python
#Creating dataframe2
df2 = pd.DataFrame({'Name':['Jeevan', 'Raavan', 'Geeta', 'Bheem'],
                    'Salary':[100000, 50000, 20000, 40000]})
df2
```

```python
#Merging two dataframes
df = pd.merge(df1, df2)
df
```

```python
import pandas as pd

# Apply function
def fun(value):
    if value > 70:
        return "Yes"
    else:
        return "No"

# Fix: The 'df' DataFrame was overwritten by the merge operation
# and no longer contains a 'bmi' column. We will use the 'data_frame'
# variable which holds the original insurance data.
# Also, 'bmi' values are not typically greater than 70, so the function
# will always return "No" for the 'bmi' column of the insurance data.
# We will create a new column 'bmi_status' to store the result.

# Reload the original insurance data to apply the function to the 'bmi' column.
# Assuming 'insurance.csv' is in the current environment.
original_insurance_df = pd.read_csv('LAB-01_insurance.csv')

original_insurance_df['bmi'] = original_insurance_df['bmi'].apply(fun)
original_insurance_df.head(10)
```
