```python
np.random.seed(42)

X = np.random.rand(1000, 20) * 100
```

```python
X
```

```python
Y = 3.5 * X + np.random.randn(X.shape[0], 1) * 20
```

```python
Y
```

```python
import pandas as pd
import numpy as np

# Ensure Y has the same number of samples as X
# If X has changed its number of samples, Y needs to be re-generated accordingly.
# Assuming Y is intended to be generated from a single feature of X, plus noise
if X.shape[0] != Y.shape[0]:
    # Use the first feature of X for generating Y, maintaining the original relationship structure
    Y = 3.5 * X[:, 0].reshape(-1, 1) + np.random.randn(X.shape[0], 1) * 20

# Now, create the DataFrame for plotting. We'll plot the first feature of X against Y.
data = {'X': X[:, 0], 'Y': Y[:, 0]}
df = pd.DataFrame(data)
df.plot(x ='X', y='Y', kind = 'scatter')
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X[:, 0].reshape(-1, 1), Y[:, 0]) # Fit the model using the first feature of X and the first feature of Y

Y_pred = model.predict(X[:, 0].reshape(-1, 1)) # Calculate predictions using the fitted model with the first feature of X

plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], Y[:, 0], color='blue', label='Data Points') # Plot the first feature of X against Y
plt.plot(X[:, 0], Y_pred, color='red', linewidth=2, label='Regression Line') # Plot the first feature of X against Y_pred
plt.title('Linear Regression on Random Dataset')
plt.xlabel('X (first feature)')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```
